## Background Story

BrightMart Retail Solutions is a fast-growing retail company in Lagos, Nigeria. The company sells electronics, groceries, and household products across multiple branches.

As the business grows, the management team has noticed several challenges:

- Customer records are not properly organized
- Staff manually calculate sales and discounts
- Daily reports take too much time to prepare
- Managers cannot quickly identify top-performing products
- The company wants to automate simple business operations using Python

You have just been hired as a **Junior Python Developer/Data Associate** at **BrightMart Retail Solutions**.

Your task is to build a **Python-based mini management system** that will help the company organize customer information, process sales, and generate simple reports.

# Project Tasks

## TASK 1 – Company Welcome Page

Create a welcome page for BrightMart Retail Solutions.

The program should:

- Display the company name
- Display a welcome message
- Ask the user to enter:
    - Staff Name
    - Branch Name
    - Current Date

**Hint:**    
**Expected Example:**
- Welcome to BrightMart Retail Solutions
- Staff Name: Collins
- Branch: Ikeja
- Date: 2026-05-16

Each customer should have:

- Customer ID

- Phone Number
- Loyalty Membership Status (`True` or `False`)

- 5 customers

- Use dictionaries
- Store all customers inside a list

## TASK 4 – Sales Processing System

The company wants to automate sales processing. 

Create a program that: 

1. Allows a customer to buy products
2. Calculates:
    - Total Price
    - Discount
    - Final Amount

**Discount Rules**

| Amount | Discount |
|---|---|
| ₦100,000 and above | 15% |
| ₦50,000 – ₦99,999 | 10% |
| Below ₦50,000 | No Discount |


**Requirements:**
- Use `if`, `elif`, and `else`
- Use arithmetic operators
- Use comparison operators


## TASK 5 – Login Verification System

Create a while loop that:

- Continuously asks for a password

**Rules:**
- Correct password: `brightmart123`
- Print:
    - `Access Granted`  for correct password
    - `Invalid Password` for wrong passwords

**Requirements:**
- Use a `while` loop

## TASK 8 – Lambda Function Challenge 

The finance department wants to quickly calculate VAT for product prices.

prices = [5000, 12000, 8000, 15000]

- Add 7.5% VAT to all prices
- Print the updated prices

**Hint:**
- **Expected Concept:**
    - updated_prices = list(map(lambda x: x * 1.075, prices))
