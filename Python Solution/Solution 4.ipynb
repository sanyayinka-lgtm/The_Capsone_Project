```python
amount = int(input("Enter the amount you want to spend: "))

if amount >= 100000:
    discount = 15 
elif amount >= 50000 and amount < 100000:
    discount = 10
else:
    discount = 0

final_amount = amount - (amount * discount / 100)
print(f"The final amount after applying the discount is: ${final_amount:.2f}")
```
