```python
customers = ["John", "Alice", "Mike", "Sarah"]

products = {
    "Laptop": 1200,
    "Phone": 800,
    "Keyboard": 100,
    "Mouse": 50
}

sales = [1200, 800, 100, 50, 800]


total_customers = len(customers)
total_products = len(products)
total_sales = sum(sales)
highest_sale = max(sales)
lowest_sale = min(sales)

print(f"Total Customers: {total_customers}")
print(f"Total Products: {total_products}")
print(f"Total Sales: ${total_sales}")
print(f"Highest Sale: ${highest_sale}")
print(f"Lowest Sale: ${lowest_sale}")

for sale in sales:
    if sale > 50:
        print(f"Sale amount: ${sale}")
for product in products:
    if products[product] > 100:
        print(f"Product: {product}, Price: ${products[product]}")
```
