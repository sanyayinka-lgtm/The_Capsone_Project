```python
transactions = [5000, 12000, -4000, 7000, 0, 25000, -2000]

for transaction in transactions:
    if transaction < 0:
        print(f"Transaction of ${-transaction} is in debt.")
        continue
    elif transaction == 25000:
        print(f"Maximum Transaction of ${transaction} is reached.")
        break
    elif transaction == 0:
        print("Transaction of $0 is neutral.")
        pass
    else:
        print(f"Transaction of ${transaction} is successful.")
```
